# Díra ve vědru - Bernoulliho rovnice 

Představte si, že máme vědro s vodou, do kterého vyřežeme malou dirku a voda touto dirkou začne stříkat. Jakou rychlost bude voda mít? A jak daleko voda dostříkne? Na to se podíváme v této úloze.   

### Rovnice kontinuity 

Představte si trubku o průřezu $S_1$, ve které teče voda (pro náš případ dokonalá kapalina - nestlačitelna) rychlostí $v_1$. Trubka se po nějaké době zůží na průřez $S_1$ a rychlost vody se zvýší na $v_2$. Protože je naše kapalina v trubce nestlačitelná, tak musí být objemový průtok trubkou konstantní. To se dá vyjádřit následující rovnicí

\begin{equation*}
    S_1 v_1 = S_2 v_2,
\end{equation*}

kterou označujeme jako rovnice kontinuity. Jedná se jinou obdobu zákona zachování hmotnosti.

<img src="kontinuita.jpg" width="500"/>

Ze zákona o zachování mechanické energie lze pro proudění doklonalé kapaliny na povrchu Země odvodit Bernoulliho rovnici 

\begin{equation*}
    \frac{1}{2}\rho v_1^2 + \rho g h_1 + p_1 = \frac{1}{2}\rho v_2^2 + \rho g h_2 + p_2,
\end{equation*}

kde $\rho$ je hustota kapaliny, $v_1$, $p_1$ a $h_1$ je rychlosti kapaliny, tlak kapaliny a výška kapaliny (vzhledem k námi zvolené výšce) v místě 1 a $v_2$, $p_2$ a $h_2$ jsou ty stejné veličiny ale v místě 2.

Pomocí těchto dvou rovnic, jsme schopní určit velikost rychlosti a místo dopadu kapaliny stříkající z otvoru vědra.

Díru do vědra jsme vyřízli ve vzdálenosti $h$ od hladiny kapaliny, tak jak je vyobrazeno na obrázku níže. Průřez této dírky je výrazně menší než průřez vědra - $S_2 \ll S_1$

<img src="images.jpeg" width="500"/>

Z rovnice kontinuity můžeme vyjadřit rychlost toku v místě 1

\begin{gather*}
    v_1 S_1 = v_2 S_2, \\
    v_1 = \frac{S_2}{S_1} v_2.
\end{gather*}

Tuto hodnotu vložíme do Bernoulliho rovnice (s tím, že si zvolíme $h_2 = 0 \text{ m}$ => $h_1 = h$) a vyjádříme si rychlost $v_2$

\begin{gather*}
    \frac{1}{2}\rho \left(\frac{S_2}{S_1} v_2\right)^2 + \rho g h + p_1 = \frac{1}{2} \rho v_2^2 + p_2, \\
    v_2^2 = \frac{\rho g h + p_1 - p_2}{\frac{1}{2} \rho \left( 1 - \frac{S_2}{S_1} \right)}
 \end{gather*}

 Díky našemu předpokladu malé dírky $S_2 \ll S_1$ můžeme druhý člen ve jmenovateli zanedbat. Jaký tlak je v místech 1 a 2? V obou techto místech je atmosferický tlak $p_a$ okolí, takže se členy $p_1$ a $p_2$ odečtou. Tím pádem si rychlost kapaliny v místě 2 můžeme jednoduše vyjádřit

 \begin{gather*}
    v_2^2 = \frac{\rho g h + p_a - p_a}{\frac{1}{2} \rho }, \\
    v_2 = \sqrt{2gh}.
\end{gather*}

Zjistili jsme tedy, že rychlost vody v místě vyříznuté dírky záleží jen na výšce hladiny nad toutu dírkou.

A jak tedy dopočítáme vzdálenost kam voda dostříkne? V okamžik, kdy voda opustí vědro naší dírkou, má rychlost jen v horizontálním směru a její velikost je $v_2$. Označme ji $v_x = v_2$. 

Na kapalinu pořád působí gravitační síla. Výška dírky nad povrchem Země je podle obrázku výše $y = H-h$. Za jak dlouho dopadne kapka vody, která právě opustila vědro dirkou, na zem? Kapka bude podstupovat rovnoměrně zrychlený pohyb. Pomocí známého vzorce pro rovnoměrně zrychlený pohyb

\begin{equation*}
    y = \frac{1}{2}at^2 + v_0 t + y_0,
\end{equation*}

můžeme dopočítat potřebný čas k dopadnutí na zem.

\begin{gather*}
    H-h = \frac{1}{2}gt^2, \\
    t = \sqrt{\frac{2}{g}(H-h)}.
\end{gather*}

Když teď známe čas pádu kapky můžeme dopočítat x-ovou vzdálenost, kterou kapka urazí. 

\begin{equation*}
    x = v_x t = \sqrt{2gh} \sqrt{\frac{2}{g}(H-h)} = 2 \sqrt{h(H-h)}
\end{equation*}

Zajimavé je, že to jak daleko voda dostříkne nezaleží na její hustotě ani na tíhovém zrychlení! Zkusme do našich vztahu nyní dosadit a vykreslit si obrázek.

Pokud budeme mít vědro o výšce $H = 1 \text{m}$, které je naplněno vodou až po okraj a uděláme do něj dírku ve výšce $h = 0.5 \text{m}$, tak rychlost vody, která právě opouští vědro dirkou je $v_x = \sqrt{2gh} = \sqrt{g} \doteq 3.13 \text{ m}$. Vzdálenost, do které voda dostříkne bude $x = 2\sqrt{h(H-h)} = 2\sqrt{0.5\cdot 0.5} \text{ m}= 1 \text{ m}$. To po tom co jsme si odvodili naše vzorce bylo jednoduché. Pojdmě si vykreslit trajektorii stříkající kapaliny. 

Jak? Musíme si vyjádřit výšku kapky $y$ v závislosti na $x$. Jak už jsme psali výše, tak kapka cestuje rovnoměrně zrychleně. Takže její výška záleží na čase podle rovnice

\begin{equation*}
    y = - \frac{1}{2} g t^2 + H-h
\end{equation*}

Místo proměnné času $t$ bychom chtěli do rovnice dosadit x-ovou souřadnici. Souřadnice $x$ je s čase provázána rovnicí 

\begin{gather*}
    x = v_x t, \\
    t = \frac{x}{v_x} = \frac{x}{\sqrt{2gh}}.
\end{gather*}

Po dosazení do rovnice pro výšku kapaliny nám vyjde závislost $y(x)$:

\begin{equation*}
    y = - \frac{1}{4h} x^2 + H - h.
\end{equation*}

Toto je rovnice trajektorie naší kapaliny. Pojdmě ji vykreslit!

```python
# Definujeme proměnné

H = 1               # [m]
h = 0.5             # [m]

# Definujeme si funkci pro parabolu

def parab(a,b,c,x):
    return(a*x**2+b*x+c)
```

```python
# Vykreslení paraboly

x = np.linspace(0,1,1000)

plt.plot(x, parab(-1/(4*h),0,H-h,x))

plt.ylabel("$y$ [m]")
plt.xlabel("$x$ [m]")
```

Podívejme se na to, jak by vypadali trajektorie pro různé hodnoty $h$

```python
# plt.plot(x, parab(-1/(4*h),0,H-h,x), label = f"$h =$ {h:,.2f} m")

h_index = np.linspace(0.1,0.9,5)

for i in range(len(h_index)):
    plt.plot(x, parab(-1/(4*h_index[i]),0,H-h_index[i],x), label = f"$h = {h_index[i]:,.2f} m$")

plt.ylim(0,1)
plt.ylabel("$y$ [m]")
plt.xlabel("$x$ [m]")

plt.legend()
```
