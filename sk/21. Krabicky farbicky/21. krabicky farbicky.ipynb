# Krabičky Farbičky - hazard

Svetlo je druh elektromagnetického žiarenia a môžeme ho rozdeliť na rôzne vlnové dĺžky, ktoré tvoria spektre žiarenia. Svetlo, ktoré vidíme, je známe ako viditeľné svetlo, a to tvorí len malú časť celkového elektromagnetického spektra. Viditeľné svetlo obsahuje rôzne farby, ktoré zodpovedajú rôznym vlnovým dĺžkam.

Spektrum viditeľného svetla ide od fialovej (najkratšia vlnová dĺžka, asi 400 nm) cez modrú, zelenú, žltú, oranžovú až po červenú (najdlhšia vlnová dĺžka, asi 700 nm).

Dnes si trošku zahazardujeme a budeme stávkovať ktorému senzoru bude najrýchlejšie stúpať teplota.

461083910_1191156631964627_5710455615470993768_n.jpg

Ku senzorom máme krabičky, ktorými ich môžeme uzavrieť. Krabičky majú rôzne farby. Všetky senzory položíme na parapetu a necháme na ne svietiť svetlo. Žiaci môžu vopred hlasovať ktorý senzor pod krabičkou ktorej farby bude mať navyššiu teplotu. Bude rozdiel v teplote vôbec? Čo sa bude diať? Dáta o teplote možno nechať bežat v pozadí a počas vyučovania sa k ním vrátiť. Senzory je vhodné si očísliť.

```python
# Nacitanie dat z merania 2
df = pd.read_excel('experiment_krabicky.xlsx')

# Vytvorime si stlpec ktory bude iba pocet sekund od zaciatku merania
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['sec'] = (df['timestamp'] - df['timestamp'].iloc[0]).dt.total_seconds()
```

```python
# Nacitanie dat z merania 1
df2 = pd.read_excel('experiment_krabicky2.xlsx')

# Vytvorime si stlpec ktory bude iba pocet sekund od zaciatku merania
df2['timestamp'] = pd.to_datetime(df2['timestamp'])
df2['sec'] = (df2['timestamp'] - df2['timestamp'].iloc[0]).dt.total_seconds()

```

```python
plt.figure(figsize=(10, 6))
plt.plot(df2['sec']/60, df2['t2.1'], label=r'$t_{hneda}$')
plt.plot(df2['sec']/60, df2['t2'], label=r'$t_{biela}$')

plt.xlabel('Čas [min]')
plt.ylabel('Teplota [C]')
plt.legend()
```

Prečo veci majú farbu?

Keď svetlo dopadne na objekt, môže byť absorbované, odrazené alebo prechádzať cez objekt. Farba objektu závisí od toho, ktoré vlnové dĺžky svetla objekt absorbuje a ktoré odráža.

Biely objekt odráža takmer všetky vlnové dĺžky viditeľného svetla, takže vidíme bielu farbu.

Čierny objekt absorbuje takmer všetko svetlo, preto nevidíme žiadnu konkrétnu farbu, iba čiernu.

Farebné objekty odrážajú iba určité vlnové dĺžky svetla a ostatné absorbujú. Napríklad, červený objekt odráža vlnovú dĺžku, ktorá zodpovedá červenej farbe (asi 700 nm), a absorbuje ostatné farby.

Farba objektu teda závisí od toho, aké svetlo odráža do našich očí. Keď vidíme napríklad modrý predmet, znamená to, že povrch objektu absorbuje všetky iné farby okrem modrej, ktorú odráža. Tento odrazený modrý svetelný signál potom vnímajú naše oči a mozog ho interpretuje ako modrú farbu. V skutočnosti sa jedná o velmi komplikované kvanotovo-mechanické javy, o ktorých sa učí neskôr na Univerzite.
