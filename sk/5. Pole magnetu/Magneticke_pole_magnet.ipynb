```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

Načítaním excelového súboru nahráme a vykreslíme zložky magnetického poľa.

Celá informácia o magnetickom poli v danom mieste je však ukrytá vo veľkosti vektora. Ak by sme chceli poznať smer magnetických siločiar a zároveň presnej informácii o smere siločiar, zaujíma nás celkový vektor. Veľkosť vektora sa dá spočítať ako

$$|\vec{B}|=\sqrt{B_x^2+B_y^2+B_z^2}$$

Úloha - spočítajte a vykreslite veľkosť magnetického poľa

```python
df = pd.read_excel(r"/content/neod.xlsx")
format = '%Y-%m-%d %H:%M:%S.%f'

# ulozenie prislusnych stlpcov do vektorov

t0 = datetime.strptime(df.iloc[0,0], format).timestamp()
time = np.array([datetime.strptime(x, format).timestamp() for x in df.iloc[:, 0] ])[0:-1] - t0
b_x = np.array(df.iloc[:, 10][0:-1] )
b_y = np.array(df.iloc[:, 11][0:-1] )
b_z = np.array(df.iloc[:, 12][0:-1] )

# veľkosť vektora magnetického poľa
b = np.sqrt(b_x**2 + b_y**2 + b_z**2)
```

```python
plt.plot(time,b_x,label='$B_x$')
plt.plot(time,b_y,label='$B_y$')
plt.plot(time,b_z,label='$B_z$')
plt.plot(time,b,label='$|B|$')
plt.xlabel('time [s]')
plt.ylabel('B [mG]')
plt.legend()
```

```python
import plotly.graph_objects as go
import numpy as np

np.random.seed(42)
random_data = np.random.randn(100)

fig = go.Figure()

fig.add_trace(go.Scatter(
    x = time,
    y = b,
    mode='lines+markers',
    name='Magnetické pole'
))

fig.update_traces(
    hoverinfo='x+y',
    hovertemplate='X: %{x}<br>Y: %{y}<extra></extra>',
    marker=dict(size=5),
    line=dict(width=1)
)

fig.update_layout(
    hovermode='y unified',
    xaxis=dict(showspikes=True),
    yaxis=dict(showspikes=True),
)

fig.show()
```

Pomocou interaktívneho grafu odčítajte namerané hodnoty veľkosti magnetického poľa pre príslušné vzdialenosti a vykreslite túto závislosť. Nezabudnite odčítať hodnotu magnetického poľa $B_0$ na pozadí.

```python
# vypocet frekvencie otacania centrifugy

B0 = 1284
L = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0] # pole, do ktoreho sa zapisu vzdialenosti senzoru od osi otacania v cm
B =  np.array([5371, 5190, 3536, 2689, 2193, 1797, 1594, 1387 ]) - B0 # pole, do ktoreho sa zapisu odcitane hodnoty magnetickeho pola
```

```python
plt.plot(L,B)

plt.ylabel('B [mG]')
plt.xlabel('L [cm]')
```
