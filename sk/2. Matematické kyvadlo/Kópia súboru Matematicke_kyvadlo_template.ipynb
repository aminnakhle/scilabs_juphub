Dobrovoľná časť.

Každé fyzikálne meranie má chybu. Jednou z mnohých metód, ako túto chybu vyjadriť je dané meranie zopakovať a chybu spočítať takzvanú smerodajnú odchylku. Nameraná hodnota má teda tvar $x_{\rm{ave}} \pm \sigma$, kde $x_{\rm{ave}}$ je aritmetický priemer a $\sigma$ je smerodadná odchylka.

Aritmetický priemer sa dá spočítať pomocou vzorca

$x_{\rm{ave}} = \sum_{n=1}^{N}\frac{1}{N}x_n.$

Smerodatná odchylka

$\sigma = \sqrt{\frac{1}{N}\sum_{n=1}^{N}(x_n-x_{\rm{ave}})}.$

Takýto zápis sa dá chápať tak, že je 68% pravdepodobnosť, že nameraná hodnota sa nachádza v intervale $[x_{\rm{ave}}-\sigma, x_{\rm{ave}}+\sigma]$.
