# Moment zotrvačnosti

Moment zotrvačnosti je fyzikálna veličina, ktorá vyjadruje mieru odporu telesa voči zmene jeho rotačného pohybu. Podobne ako hmotnosť určuje odpor telesa voči zmene pohybu pri priamočiarom pohybe, moment zotrvačnosti závisí od rozloženia hmoty v tele vzhľadom na os rotácie. Telesá s väčším momentom zotrvačnosti sa ťažšie roztočia alebo zastavia. Moment zotrvačnosti je kľúčový pri opisovaní rotácie, pretože určuje, aký veľký krútiaci moment je potrebný na zmenu uhlovej rýchlosti telesa. V praxi to znamená, že telesá s hmotou sústredenou ďalej od osi rotácie majú väčší moment zotrvačnosti a tým pádom väčšiu stabilitu pri rotácii.

Pre momenty zotrvačnosti telesa v tvare kvádru voči osiam x y z ako je znázornené na obrázku platia nasledujúce vzťahy.
$$
    I_1 = \frac{1}{12} m (b^2 + c^2), \quad I_2 = \frac{1}{12} m (a^2 + c^2), \quad I_3 = \frac{1}{12} m (a^2 + b^2).
$$

kde a b c sú dĺžky hran k osiam x y z.

cuboid.png

Keď teleso vyhodime do vzduchu s počiatočnou rotáciou okolo osi x alebo z. Resp. okolo osi, v ktorom smere je najdlhši alebo najkratší rozmer kvádru, môžeme zložky uhlovej rychlosti v osi spojenej s telesom popísať následovne napríklad následovne pomocou tajuplných konštánt α a β, v ktorých sú zakódované momenty zotrvačnosti:

$$
\omega_1(t) = košt.,
$$

$$
\omega_2(t) = \sqrt{\alpha}~K\cos(\sqrt{\alpha\beta}~\omega_1 t + \varphi),
$$

$$
\omega_3(t) = -\sqrt{\beta~}K\sin(\sqrt{\alpha\beta}~\omega_1 t + \varphi).
$$

V rámci tohto cvičenia, a asi aj štúdia na strendej škole, bude vhodné keď si len uvedomíme, že uhlová rýchlosť v osi v ktorej sme kváder hodili je konštantná a vo zvyšných dvoch bude popísana dvoma goniometrickými funkciami. Zameráme sa na ne.

Plné znenie týchto rovníc je mimo osnovy. Čo by sme vedeli ale odmerať sú amplitúdy týchto funkcii a ich pomer. Z neho by sme dekódovali následoné:
$$
\sqrt{\frac{\alpha}{\beta}}=\sqrt{\frac{{I_3}^2 - I_1I_3}{{I_2}^2 - I_1I_2}}.
$$

Najjednoduchšie sa vo Fyzike merajú pomery, keďže takým spôsobom eliminujeme prácu s jednotkami, potencialne zlé prevedenia a iné chyby.

Tak ako v iných úlohach sme už použili, využijeme tvorbu interaktívneho grafu plotly, z ktorého budme môct hodnoty odčítať pohybom kurzoru po grafe.

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=df_subset['sec'], y=df_subset['gyroscope.x'], mode='lines', name=r'$\omega_{1}$'))
fig.add_trace(go.Scatter(x=df_subset['sec'], y=df_subset['gyroscope.y'], mode='lines', name=r'$\omega_{2}$'))
fig.add_trace(go.Scatter(x=df_subset['sec'], y=df_subset['gyroscope.z'], mode='lines', name=r'$\omega_{3}$'))

fig.update_layout(
    xaxis_title="Čas [min]",
    yaxis_title="Uhlová rýchlosť [°/s]",
    legend_title="Legend"
)

fig.show()
```
