# Diera vo vedre - Bernoulliho rovnica

Predstavte si, že máme vedro s vodou, do ktorého vyrežeme malú dierku a voda touto dierkou začne striekať. Akú rýchlosť bude voda mať? A ako ďaleko voda dostrekne? Na to sa pozrieme v tejto úlohe.

### Rovnica kontinuity

Predstavte si rúrku s prierezom $S_1$, v ktorej tečie voda (pre náš prípad dokonalá kvapalina - nestlačiteľná) rýchlosťou $v_1$. Rúrka sa po nejakom čase zúži na prierez $S_1$ a rýchlosť vody sa zvýši na $v_2$. Pretože je naša kvapalina v rúrke nestlačiteľná, tak musí byť objemový prietok rúrkou konštantný. To sa dá vyjadriť nasledujúcou rovnicou

\begin{equation*}
 S_1 v_1 = S_2 v_2,
\end{equation*}

ktorú označujeme ako rovnica kontinuity. Jedná sa inú obdobu zákona zachovania hmotnosti.


kontinuita.jpg

Zo zákona o zachovaní mechanickej energie je možné pre prúdenie doklonalej kvapaliny na povrchu Zeme odvodiť Bernoulliho rovnicu

\begin{equation*}
 \frac{1}{2}\rho v_1^2 + \rho g h_1 + p_1 = \frac{1}{2}\rho v_2^2 + \rho g h_2 + p_2,
\end{equation*}

kde $\rho$ je hustota kvapaliny, $v_1$, $p_1$ a $h_1$ je rýchlosť kvapaliny, tlak kvapaliny a výška kvapaliny (vzhľadom k nami zvolenej výške) v mieste 1 a $v_2$, $p_2$ a $h_2 $ sú tie isté veličiny ale v mieste 2.

Pomocou týchto dvoch rovníc sme schopní určiť veľkosť rýchlosti a miesto vplyvu kvapaliny striekajúcej z otvoru vedra.

Dieru do vedra sme vyrezali vo vzdialenosti $h$ od hladiny kvapaliny, tak ako je vyobrazené na obrázku nižšie. Prierez tejto dierky je výrazne menší ako prierez vedra - $S_2 \ll S_1$

Z rovnice kontinuity môžeme vyjadriť rýchlosť toku v mieste 1

\begin{gather*}
 v_1 S_1 = v_2 S_2, \\
 v_1 = \frac{S_2}{S_1} v_2.
\end{gather*}

Túto hodnotu vložíme do Bernoulliho rovnice (s tým, že si zvolíme $h_2 = 0 \text{ m}$ => $h_1 = h$) a vyjadríme si rýchlosť $v_2$

\begin{gather*}
 \frac{1}{2}\rho \left(\frac{S_2}{S_1} v_2\right)^2 + \rho g h + p_1 = \frac{1}{2} \rho v_2^2 + p_2, \\
 v_2^2 = \frac{\rho g h + p_1 - p_2}{\frac{1}{2} \rho \left( 1 - \frac{S_2}{S_1} \right)}
 \end{gather*}

 Vďaka nášmu predpokladu malej dierky $S_2 \ll S_1$ môžeme druhý člen v menovateľovi zanedbať. Aký tlak je v miestach 1 a 2? V oboch týchto miestach je atmosférický tlak $p_a$ okolia, takže sa členmi $p_1$ a $p_2$ odčítajú. Tým pádom si rýchlosť kvapaliny v mieste 2 môžeme jednoducho vyjadriť

 \begin{gather*}
 v_2^2 = \frac{\rho g h + p_a - p_a}{\frac{1}{2} \rho }, \\
 v_2 = \sqrt{2gh}.
\end{gather*}

Zistili sme teda, že rýchlosť vody v mieste vyrezanej dierky záleží len na výške hladiny nad touto dierkou.

A ako teda dopočítame vzdialenosť kam voda dostrekne? V okamihu, keď voda opustí vedro našou dierkou, má rýchlosť len v horizontálnom smere a jej veľkosť je $v_2$. Označme ju $v_x = v_2$.

Na kvapalinu stále pôsobí gravitačná sila. Výška dierky nad povrchom Zeme je podľa obrázku vyššie $y = H-h$. Za ako dlho dopadne kvapka vody, ktorá práve opustila vedro dirkou, na zem? Kvapka bude podstupovať rovnomerne zrýchlený pohyb. Pomocou známeho vzorca pre rovnomerne zrýchlený pohyb

\begin{equation*}
 y = \frac{1}{2}at^2 + v_0 t + y_0,
\end{equation*}

môžeme dopočítať potrebný čas na dopadnutie na zem.

\begin{gather*}
 H-h = \frac{1}{2}gt^2, \\
 t = \sqrt{\frac{2}{g}(H-h)}.
\end{gather*}

Keď teraz poznáme čas pádu kvapky môžeme dopočítať x-ovú vzdialenosť, ktorú kvapka prejde.

\begin{equation*}
 x = v_x t = \sqrt{2gh} \sqrt{\frac{2}{g}(H-h)} = 2 \sqrt{h(H-h)}
\end{equation*}

Zaujimavé je, že to ako ďaleko voda dostrekne, nezaleží na jej hustote ani na tiažovom zrýchlení! Skúsme do našich vzťahov teraz dosadiť a vykresliť si obrázok.

Ak budeme mať vedro s výškou $H = 1 \text{m}$, ktoré je naplnené vodou až po okraj, a urobíme doň dierku vo výške $h = 0.5 \text{m}$, tak rýchlosť vody, ktorá práve opúšťa vedro dirkou, je $v_x = \sqrt{2gh} = \sqrt{g} \doteq 3.13 \text{ m}$. Vzdialenosť, do ktorej voda dostrekne, bude $x = 2\sqrt{h(H-h)} = 2\sqrt{0.5\cdot 0.5} \text{ m}= 1 \text{ m}$. To po tom, čo sme si odvodili naše vzorce, bolo jednoduché. Poďme si vykresliť trajektóriu striekajúcej kvapaliny.

Ako? Musíme si vyjadriť výšku kvapky $y$ v závislosti od $x$. Ako sme už písali vyššie, tak kvapka cestuje rovnomerne zrýchlene. Takže jej výška záleží na čase podľa rovnice

\begin{equation*}
 y = - \frac{1}{2} g t^2 + H-h
\end{equation*}

Namiesto premennej času $t$ by sme chceli do rovnice dosadiť x-ovú súradnicu. Súradnica $x$ je s čase previazaná rovnicou

\begin{gather*}
 x = v_x t, \\
 t = \frac{x}{v_x} = \frac{x}{\sqrt{2gh}}.
\end{gather*}

Po dosadení do rovnice pre výšku kvapaliny nám vyjde závislosť $y(x)$:

\begin{equation*}
 y = - \frac{1}{4h} x^2 + H - h.
\end{equation*}

Toto je rovnica trajektórie našej kvapaliny. Poďme ju vykresliť!

```python
# Definujeme premenné

H = 1 # [m]
h = 0.5 # [m]

# Definujeme si funkciu pre parabolu

def parab(a,b,c,x):
 return(a*x**2+b*x+c)
```

```python
# Vykreslenie paraboly

x = np.linspace(0,1,1000)

plt.plot(x, parab(-1/(4*h),0,H-h,x))

plt.ylabel("$y$ [m]")
plt.xlabel("$x$ [m]")
```

Pozrime sa na to, ako by vyzerali trajektórie pre rôzne hodnoty $h$

```python
# plt.plot(x, parab(-1/(4*h),0,H-h,x), label = f"$h =$ {h:,.2f} m")

h_index = np.linspace(0.1,0.9,5)

for i in range(len(h_index)):
    plt.plot(x, parab(-1/(4*h_index[i]),0,H-h_index[i],x), label = f"$h = {h_index[i]:,.2f} m$")

plt.ylim(0,1)
plt.ylabel("$y$ [m]")
plt.xlabel("$x$ [m]")

plt.legend()
```
