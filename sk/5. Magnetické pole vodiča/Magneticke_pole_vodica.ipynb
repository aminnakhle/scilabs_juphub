**Magnetické pole vodiča**

Vodiče vedú elektrický prúd vďaka voľným elektrónom, ktoré sa môžu voľne pohybovať po celom materiáli. V kovoch sú valenčné elektróny slabo viazané k atómom a tvoria elektrónový plyn. Keď na vodič zapojíme pod elektrické napätie, tieto elektróny sa usporiadane pohybujú jedným smerom, čo tvorí elektrický prúd.

Pomocou experimentu môžme vyskúšať, aký má elektrický prúd vplyv na magnetické pole v okolí vodiča. To sa dá zapojením jednoduchého obvodu na obrázku a položením senzoru na vodič.

```python
# nacitanie kniznic
from google.colab import files
from google.colab import drive

from datetime import datetime
import pandas as pd
import numpy as np
import scipy as sp
from scipy.optimize import curve_fit
#from scipy.optimize import integrate
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"/content/vodic_on_off.xlsx") # nacitanie excelovskych dat
format = '%Y-%m-%d %H:%M:%S.%f'

# ulozenie prislusnych stlpcov do vektorov

t0 = datetime.strptime(df.iloc[0,0], format).timestamp()
time = np.array([datetime.strptime(x, format).timestamp() for x in df.iloc[:, 0] ])[0:-1] - t0
b_x = np.array(df.iloc[:, 10][0:-1] )
b_y = np.array(df.iloc[:, 11][0:-1] )
b_z = np.array(df.iloc[:, 12][0:-1] )
b = np.sqrt(b_x**2+b_y**2+b_z**2)
```

```python
plt.plot(time,b)
plt.xlabel('Time [s]')
plt.ylabel('B [mG]')
```

Hodnota magnetického poľa na začiatku odpovedá veľkosti pozadia (magnetické pole Zeme, elektronika v okolí...). Po zapnutí spínača vidíme, že hodnota magnetického poľa narástla. To sa udialo kvôli usporiadanému pohybu náboja vo vodiči, alebo inak povedané, prúdu. Po prerušení obvodu organizovaný pohyb elektrónov ustál a magnetické pole sa opäť vrátilo na pôvodnú hodnotu.
