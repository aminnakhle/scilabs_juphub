**Vektor**

Ak sa teleso nachádza v gravitačnom poli Zeme, pôsobí naň gravitačná sila o veľkosti približne G = 9.8 $\rm{m/s^2}$. Senzor SciLabs je normovaný práve touto hodnotou a senzor v kľude meria hodnotu 1G. Vektor však meria zri zložky zrýchlenia, ktoré sú závislé od natočenia senzoru. Napríklad, vektor gravitačného zrýchlenia $\vec{g}$ natočený podľa obrázka sa dá vyjdriť v súradnicovej sústave senzoru ako $(0,0,-1)$. Ak by sme senzor otočili o 180 stupňov, namerali by sme hodnotu $(0,0,1)$

Úloha 1: Postavte na krátky úsek (pár sekúnd) senzor na každú zo šiestich hrán a pozorujte, ako vyzerá vektor zrýchlenia. Zakreslite, ako vyzerá súradnicová os spojená so senzorom.

```python
# nacitanie kniznic
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_excel(r"/content/hrany.xlsx") # nacitanie excelovskych dat
format = '%Y-%m-%d %H:%M:%S.%f'

# ulozenie prislusnych stlpcov do vektorov

t0 = datetime.strptime(df.iloc[0,0], format).timestamp()
time = np.array([datetime.strptime(x, format).timestamp() for x in df.iloc[:, 0] ])[0:-1] - t0
acc_x = np.array(df.iloc[:, 1][0:-1] )
acc_y = np.array(df.iloc[:, 2][0:-1] )
acc_z = np.array(df.iloc[:, 3][0:-1] )
```

```python
# vykreslenie dat

plt.plot(time,acc_x, label= '$a_x$')
plt.plot(time,acc_y, label= '$a_y$')
plt.plot(time,acc_z, label= '$a_z$')
plt.xlabel('time [s]')
plt.ylabel('a [G]')
plt.legend()
```

Z grafu vidno, že ak je senzor v kľude na rovnej podložke, tak hodnotu 1 ukazuje iba jedna zložka. To znamená, že zvolená podložka je rovnobežná so zemou a súradnicová os senzora je rovnobežná s hranami senzora. Otočenie senzoru o 180 stupňov spôsobilo, že nameraná hodnota zmení znamienko, otočenie na inú hranu spôsobilo že bola nenulová iná zložka zrychlenia.
